#### Basic Histogram 

```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
```

```python
marks = np.array([56,66,56,87,67,62,28,65,48,89,82,66,59,52,60,63,46,67,88])
```

```python
plt.title('Histogram of Marks')
plt.xlabel('Range of Marks')
plt.xticks(range(0,101,10))
plt.grid()
plt.ylabel('Count')
plt.hist(marks, bins = range(0,101,10), color='gray');
```

```python
import seaborn as sns
```

```python
sns.histplot(marks, kde=True);
```

```python
img = cv2.imread('images/lenna.png')
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
```

```python
# images, channels, mask, histSize, range
hist = cv2.calcHist([gray], [0], None, [256], [0, 256])
```

```python
plt.title('Image Histogram')
plt.xlabel('Pixel Intensity')
plt.ylabel('Number of pixels')
plt.bar(list(range(0,256)), hist.flatten())
plt.xlim([0, 256])    # Ensure correct x-axis range
plt.show()
```

```python
plt.figure(figsize=(16,9))
plt.title('Image Histogram')
plt.xlabel('Pixel Intensity')
plt.ylabel('Number of Pixels')
plt.xticks(range(0,256,10))
plt.hist(gray.flatten(), bins=range(0,256));
```

### Color Image Histogram

```python
img = cv2.imread('images/lenna.png')
```

```python
b, g, r = cv2.split(img)
```

```python
plt.figure(figsize=(16,9))
plt.title('Color Image Histogram')
plt.subplot(1,3,1)
plt.xticks(range(0,256,20))
plt.hist(b.flatten(), bins=range(0,256), color = 'b');
plt.subplot(1,3,2)
plt.xticks(range(0,256,20))
plt.hist(g.flatten(), bins=range(0,256), color = 'g');
plt.subplot(1,3,3)
plt.xticks(range(0,256,20))
plt.hist(r.flatten(), bins=range(0,256), color = 'r');
```

####  Stretching and Shrinking

```python
image = cv2.imread('images/lenna.png')

new_width, new_height = 512, 200   # Define new dimensions (larger than original)

stretched_image = cv2.resize(image, dsize=(new_width, new_height),interpolation=cv2.INTER_CUBIC)

cv2.imshow('Stretched Image', stretched_image)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
image = cv2.imread('images/lenna.png')

new_width, new_height = 200, 512   # Define new dimensions (larger than original)

# stretch the image using interpolation
stretched_image = cv2.resize(image, dsize=(new_width, new_height),interpolation=cv2.INTER_CUBIC)

cv2.imshow('Stretched Image', stretched_image)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

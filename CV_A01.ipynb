```python
img_random = np.zeros([5, 5])
```

```python
for i in range(5):
    for j in range(5):
        img_random[i, j] = np.random.randint(0, 255)
```

```python
plt.figure(figsize=(12,6))
plt.subplot(2, 4, 1)
plt.title('Coolwarm')
plt.imshow(img_random, cmap = 'coolwarm')
plt.subplot(2, 4, 2)
plt.title('Rainbow')
plt.imshow(img_random, cmap = 'rainbow')
plt.subplot(2, 4, 3)
plt.title('Tab10')
plt.imshow(img_random, cmap = 'tab10')
plt.subplot(2, 4, 4)
plt.title('Seismic')
plt.imshow(img_random, cmap = 'seismic')
plt.subplot(2, 4, 5)
plt.title('Twilight')
plt.imshow(img_random, cmap = 'twilight')
plt.subplot(2, 4, 6)
plt.title('Turbo')
plt.imshow(img_random, cmap = 'turbo')
plt.subplot(2, 4, 7)
plt.title('BuGn')
plt.imshow(img_random, cmap = 'BuGn')
plt.tight_layout()
```

```python
vid = cv2.VideoCapture('images/nature.mp4')
```

```python
height = int(vid.get(cv2.CAP_PROP_FRAME_HEIGHT))
width = int(vid.get(cv2.CAP_PROP_FRAME_WIDTH))
frames = int(vid.get(cv2.CAP_PROP_FRAME_COUNT))
fps = int(vid.get(cv2.CAP_PROP_FPS))
```

```python
print('Height                 :', height)
print('Width                  :', width)
print('Total Frames           :', frames)
print('Frames per second      :', fps)
```

```python
while True:
    ret, frame = vid.read()
    if not ret:
        break
    cv2.imshow('Video', frame)
    if cv2.waitKey(5) == ord('q'):
        break
cv2.destroyAllWindows()
```

```python
timestamp = '00:00:50.00'
hours, minutes, seconds = [float(i) for i in timestamp.split(':')]

trigger_frame = hours * 3600 * fps + minutes * 60 * fps + seconds * fps

vid.set(1, trigger_frame)
success, frame = vid.read()
if success:
    cv2.imwrite('nature_50s.jpg', frame)
```

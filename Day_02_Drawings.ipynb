#### Draw a 3D line

```python
from mpl_toolkits.mplot3d import Axes3D
import matplotlib.pyplot as plt

start_point = (10, 22, 33)
end_point = (44, 52, 67)

fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

ax.plot([start_point[0], end_point[0]],
        [start_point[1], end_point[1]],
        [start_point[2], end_point[2]], 'r-')

ax.plot([10, 40], [55, 35], [35, 30], 'g')

ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')
ax.set_title('3D Line')
plt.show()
```

#### Rectangle using Matplotlib

```python
import matplotlib.pyplot as plt

x_min, y_min, x_max, y_max = 0, 0, 60, 70

rectangle = plt.Rectangle((x_min, y_min), x_max-x_min, y_max-y_min, color='r', alpha=0.1)

plt.plot([x_min, x_max], [y_min, y_min], 'b-', label='X-axis')
plt.plot([x_min, x_min], [y_min, y_max], 'g-', label='Y-axis')
plt.gca().add_patch(rectangle)

plt.xlabel('X')
plt.ylabel('Y')
plt.title('Rectangle plot')
plt.legend()

plt.show()
```

### Rectangle using turtle

```python
# Create a Turtle object
t = turtle.Turtle()

# set color and pen size
t.color("green")
t.pensize(5)
t.speed(0)
# Move the Turtle and draw the rectangle
t.penup()
t.goto(-50, 25)
t.pendown()
t.forward(100)
t.left(90)
t.forward(50)
t.left(90)
t.forward(100)
t.left(90)
t.forward(50)
t.hideturtle()

# Start the Turtle graphics window
turtle.done()
```

### Drawing the Ellipse

```python
# Load an image
image = cv2.imread("images/lenna.png")

# Define center coordinates radii and angle
center = (256, 256)  # (X, Y)
axes = (255, 168)     # (major_axes and minor_axes)
angle = 90     # Angle of rotation in degrees

cv2.ellipse(image, center, axes, angle, 0, 360, (0, 255, 255), 10)

cv2.imshow("image with ellipse", image)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
# Load an image
image = cv2.imread("images/lenna.png")

# Define center coordinates radii and angle
center = (256, 256)  # (X, Y)
axes = (168, 255)     # (major_axes and minor_axes)
angle = 0     # Angle of rotation in degrees

cv2.ellipse(image, center, axes, angle, 180, 360, (255, 0, 0), 10)
cv2.ellipse(image, center, axes, angle, 0, 180, (0, 0, 255), 10)

cv2.imshow("image with ellipse", image)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

### Ellipse using matplotlib

```python
import matplotlib.pyplot as plt
from matplotlib.patches import Ellipse

# Define center coordinates, radii and angle
center = (0, 0)
width = 8
height = 5
angle = 30

# Create the ellipse object
ellipse = Ellipse(xy=center, width=width, height=height, angle=angle, color='red')

# Add the ellipse to the plot
fig, ax = plt.subplots()
ax.add_patch(ellipse)
ax.set_aspect('equal')      # maintain equal aspect ratio for correct shape

# Set limits slightly wider than the ellipse
ax.set_xlim(-width - width/5, width + width/5)
ax.set_ylim(-height - height/5, height + height/5)

# Add labels and title
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_title('Ellipse plot')

# Show the plot
plt.show()
```

```python
# Load an image
image = cv2.imread("images/lenna.png")

# Define center coordinates radii and angle
center = (256, 256)

for radius in range(10, 400, 5):
    cv2.circle(image, center, radius, (0, radius//2, radius), 2)

cv2.imshow("image with Circle", image)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

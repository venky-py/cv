### Averaging Kernel

```python
import cv2
import numpy as np

image = cv2.imread('images/lenna.png')

kernel = np.ones((5,5), np.float32) / 25

smoothed_image = cv2.filter2D(image, -1, kernel)

cv2.imshow('Original', image)
cv2.imshow('Smoothed', smoothed_image)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
kernel
```

```python
final_image = np.vstack([image, smoothed_image])
```

```python
cv2.imshow('Final', final_image)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
image = cv2.imread('images/lenna.png')
kernel = np.ones((7,7), np.float32) / 49
smoothed_image = cv2.filter2D(image, -1, kernel)
final_image = np.hstack([image, smoothed_image])
cv2.imshow('Smoothed', final_image)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

#### Gaussian Kernel

```python
kernel_size = 7 
sigma = 0.05 * ((kernel_size -1) * 0.5 -1) + 2.35
kernel = cv2.getGaussianKernel(kernel_size, sigma)
```

```python
kernel
```

```python
import matplotlib.pyplot as plt
```

```python
plt.plot(range(7), kernel)
```

```python
smoothed_image = cv2.filter2D(image, -1, kernel)

cv2.imshow('Original',image)
cv2.imshow('Smoothed', smoothed_image)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
final_image = np.vstack([image, smoothed_image])

cv2.imshow('Final', final_image)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
# averaging kernel
avg_kernel = np.ones((5,5), np.float32) /25

# gaussian kernel
sigma = 0.05 * ((7 -1) * 0.5 -1) + 2.35
gauss_kernel = cv2.getGaussianKernel(7, sigma)
```

```python
avg_image = cv2.filter2D(image, -1, avg_kernel)
gauss_image = cv2.filter2D(image, -1, gauss_kernel)
```

```python
comp_image = np.hstack([image, avg_image, gauss_image])
```

```python
cv2.imshow('Final', comp_image)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

#### Median Filter

```python
img = cv2.imread('images/lenna.png')
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
```

```python
kernel_size = 15
median = cv2.medianBlur(gray, kernel_size)
```

```python
median.shape
```

```python
plt.figure(figsize=(12,9))
plt.subplot(1,2,1)
plt.title('Original Image')
plt.imshow(gray, cmap = 'gray')
plt.subplot(1,2,2)
plt.title('Median Blurred Image')
plt.imshow(median, cmap = 'gray')
```

```python
median3 = cv2.medianBlur(gray, 3)
median5 = cv2.medianBlur(gray, 5)
median7 = cv2.medianBlur(gray, 7)
median9 = cv2.medianBlur(gray, 9)
median11 = cv2.medianBlur(gray, 11)
median13 = cv2.medianBlur(gray, 13)
median15 = cv2.medianBlur(gray, 15)
median31 = cv2.medianBlur(gray, 31)
```

```python
plt.figure(figsize=(15,6))
plt.subplot(2,4,1)
plt.title('Original')
plt.imshow(gray, cmap = 'gray')
plt.subplot(2,4,2)
plt.title('3-Median')
plt.imshow(median3, cmap = 'gray')
plt.subplot(2,4,3)
plt.title('5-Median')
plt.imshow(median5, cmap = 'gray')
plt.subplot(2,4,4)
plt.title('7-Median')
plt.imshow(median7, cmap = 'gray')
plt.subplot(2,4,5)
plt.title('9-Median')
plt.imshow(median9, cmap = 'gray')
plt.subplot(2,4,6)
plt.title('11-Median')
plt.imshow(median11, cmap = 'gray')
plt.subplot(2,4,7)
plt.title('13-Median')
plt.imshow(median13, cmap = 'gray')
plt.subplot(2,4,8)
plt.title('31-Median')
plt.imshow(median31, cmap = 'gray')
plt.tight_layout()
```

### Bilateral Filter

```python
bilateral = cv2.bilateralFilter(gray, d=9, sigmaColor=95, sigmaSpace=75)

plt.figure(figsize=(10,6))
plt.subplot(121)
plt.title('Original')
plt.imshow(gray, cmap = 'gray')
plt.subplot(122)
plt.title('Bilateral Filtered Image')
plt.imshow(bilateral, cmap = 'gray')
plt.show()
```

### All Together

```python
avg_image = cv2.cvtColor(avg_image, cv2.COLOR_BGR2GRAY)
gauss_image = cv2.cvtColor(gauss_image, cv2.COLOR_BGR2GRAY)
```

```python
plt.figure(figsize=(16,8))
plt.subplot(151)
plt.title('Original')
plt.imshow(gray, cmap = 'gray')
plt.subplot(152)
plt.title('Bilateral Filtered Image')
plt.imshow(bilateral, cmap = 'gray')
plt.subplot(153)
plt.title('15-Median')
plt.imshow(median15, cmap = 'gray')
plt.subplot(154)
plt.title('Averaging')
plt.imshow(avg_image, cmap = 'gray')
plt.subplot(155)
plt.title('Gaussian')
plt.imshow(gauss_image, cmap = 'gray')
plt.show()
plt.tight_layout()
```

```python
h_image1 = np.hstack([avg_image, gauss_image])
h_image2 = np.hstack([median15, bilateral])
final_image = np.vstack([h_image1, h_image2])
plt.imshow(final_image, cmap='gray')
```

```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
```

```python
# img = cv2.imread('images/lenna.png', 0)
img = cv2.imread('images/lenna.png')
img = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
```

```python
binary_image = img > 136
```

```python
binary_image
```

```python
plt.imshow(binary_image, cmap = 'gray')
```

```python
# applying different thresholding techniques on the input image
# all pixels value above 120 will be set to 255
ret, thresh1 = cv2.threshold(img, 120, 255, cv2.THRESH_BINARY)
ret, thresh2 = cv2.threshold(img, 120, 255, cv2.THRESH_BINARY_INV)
ret, thresh3 = cv2.threshold(img, 120, 255, cv2.THRESH_TRUNC)
ret, thresh4 = cv2.threshold(img, 120, 255, cv2.THRESH_TOZERO)
ret, thresh5 = cv2.threshold(img, 120, 255, cv2.THRESH_TOZERO_INV)

plt.figure(figsize=(16,8))
plt.subplot(1, 5, 1)
plt.title("Binary Thresholding")
plt.imshow(thresh1, cmap = 'gray')
plt.subplot(1, 5, 2)
plt.title("Inverse Binary Thresholding")
plt.imshow(thresh2, cmap = 'gray')
plt.subplot(1, 5, 3)
plt.title("Truncation Thresholding")
plt.imshow(thresh3, cmap = 'gray')
plt.subplot(1, 5, 4)
plt.title("Zero Thresholding")
plt.imshow(thresh4, cmap = 'gray')
plt.subplot(1, 5, 5)
plt.title("Inverse Zero Thresholding")
plt.imshow(thresh5, cmap = 'gray')
```

```python
import cv2
import numpy as np

image = cv2.imread('images/sun.png', 0)

flip = cv2.flip(image, 0)
# final_image = np.vstack([image, flip])

resized = cv2.resize(image, None, fx=1.5, fy=1.5)

ret, thresh = cv2.threshold(resized, 120, 255, cv2.THRESH_BINARY)
cv2.imwrite('binary_image_sun.jpg', thresh)

cv2.imshow("Binary", thresh)
cv2.imshow("Resized", resized)
cv2.imshow("Flip", flip)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

#### Otsu's Thresholding

```python
import cv2
import numpy as np

img = cv2.imread('images/lenna.png')

gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

_, thresh_otsu = cv2.threshold(gray, 0, 255, cv2.THRESH_OTSU)

cv2.imshow("Original Image", img)
cv2.imshow("Otsu Thresholding", thresh_otsu)

cv2.waitKey(0)
cv2.destroyAllWindows()
```

#### Otsu's Thresholding using skimage

```python
from skimage import io, filters
image = io.imread('images/lenna.png', as_gray=True)
```

```python
threshold = filters.threshold_otsu(image)
binary_image = image > threshold
```
